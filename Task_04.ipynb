```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Bar Chart - classic
plt.style.use('classic')
students = ['Jagdish', 'Jayesh', 'Aditya', 'Rohit']
scores = [65, 72, 58, 80]
plt.bar(students, scores, color='orange')
plt.title("Bar Chart - Students' Scores")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()
```

```python
# 2. Line Chart - seaborn-v0_8-bright
plt.style.use('seaborn-v0_8-bright')
x = [1, 2, 3, 4]
y = [70, 75, 80, 85]
plt.plot(x, y, marker='o', linestyle='--')
plt.title("Line Chart - Progress")
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter Plot - ggplot
plt.style.use('ggplot')
x = [160, 165, 170, 175]
y = [50, 55, 65, 72]
plt.scatter(x, y, color='green')
plt.title("Scatter - Height vs Weight")
plt.tight_layout()
plt.show()
```

```python
# 4. Pie Chart - bmh
plt.style.use('bmh')
labels = ['Jagdish', 'Jayesh', 'Aditya', 'Rohit']
time_spent = [4, 3, 2, 5]
plt.pie(time_spent, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart - Study Time")
plt.tight_layout()
plt.show()
```

```python
# 5. Horizontal Bar Chart - fivethirtyeight
plt.style.use('fivethirtyeight')
projects = ['Jagdish', 'Jayesh', 'Aditya', 'Rohit']
completed = [3, 5, 4, 6]
plt.barh(projects, completed)
plt.title("Projects Completed")
plt.tight_layout()
plt.show()
```

```python
# 6. Stackplot - seaborn-v0_8-colorblind
plt.style.use('seaborn-v0_8-colorblind')
days = ['Mon', 'Tue', 'Wed', 'Thu']
jagdish = [2, 3, 2, 4]
jayesh = [1, 1, 2, 1]
aditya = [1, 2, 2, 1]
plt.stackplot(days, jagdish, jayesh, aditya, labels=['Jagdish', 'Jayesh', 'Aditya'])
plt.title("Study Hours per Day")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 7. Area Plot - Solarize_Light2
plt.style.use('Solarize_Light2')
months = ['Jan', 'Feb', 'Mar', 'Apr']
jayesh_sales = [20, 25, 30, 28]
rohit_sales = [15, 18, 22, 20]
plt.fill_between(months, jayesh_sales, color='blue', alpha=0.6, label='Jayesh')
plt.fill_between(months, rohit_sales, color='red', alpha=0.6, label='Rohit')
plt.title("Monthly Sales")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 8. Histogram - grayscale
plt.style.use('grayscale')
marks = np.random.randint(50, 100, 100)
plt.hist(marks, bins=10, edgecolor='black')
plt.title("Histogram - Marks Distribution")
plt.tight_layout()
plt.show()
```

```python
# 9. Step Plot - tableau-colorblind10
plt.style.use('tableau-colorblind10')
sem = ['Sem1', 'Sem2', 'Sem3', 'Sem4']
rohit_percent = [65, 70, 68, 75]
plt.step(sem, rohit_percent, where='mid', linewidth=2, marker='o')
plt.title("Rohit's Academic Progress")
plt.tight_layout()
plt.show()
```

```python
# 10. Line Plot - dark_background
plt.style.use('dark_background')
x = [0, 1, 2, 3]
y = [75, 78, 80, 85]
plt.plot(x, y, color='cyan', marker='s', linestyle='-')
plt.title("Performance Trend")
plt.tight_layout()
plt.show()
```
