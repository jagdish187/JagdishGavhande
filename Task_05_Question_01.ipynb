1. Side-by-Side Bar Chart – Comparing Runs of Sachin, Kohli, and Sehwag

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ['2005', '2007', '2009', '2011', '2013']
sachin_runs = [1200, 1000, 1100, 950, 1050]
kohli_runs = [0, 500, 900, 1300, 1150]
sehwag_runs = [850, 870, 1120, 1000, 750]

# Positioning bars
positions = np.arange(len(years))
bar_width = 0.25

# Plot
plt.bar(positions - bar_width, sachin_runs, width=bar_width, label='Sachin', color='navy')
plt.bar(positions, kohli_runs, width=bar_width, label='Kohli', color='limegreen')
plt.bar(positions + bar_width, sehwag_runs, width=bar_width, label='Sehwag', color='tomato')

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Runs Scored by Players (2005–2013)")
plt.xticks(positions, years)
plt.legend()
plt.tight_layout()
plt.show()
```

2. Horizontal Bar Chart – Total Runs in Debut 5 Years


```python
# Total runs in first 5 years of career
players = ['Sachin', 'Kohli', 'Sehwag']
debut_runs = [5400, 4900, 4600]

plt.barh(players, debut_runs, color=['orange', 'black', 'green'])
plt.xlabel("Total Runs")
plt.title("Runs in First 5 Years of Career")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
