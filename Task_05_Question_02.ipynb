1. Pie Chart – Mobile OS Market Share

```python
import matplotlib.pyplot as plt

# Market share data
labels = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2]
colors = ['green', 'red', 'gray']

# Pie chart
plt.pie(market_share, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Market Share of Mobile Operating Systems")
plt.axis('equal')  # Ensures pie is circular
plt.tight_layout()
plt.show()
```

2. Horizontal Bar Chart – Mobile OS Market Share

```python
# Horizontal bar chart
plt.barh(labels, market_share, color=colors)
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
